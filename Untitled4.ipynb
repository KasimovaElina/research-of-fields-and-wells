1.Воронка депрессии.(нестационарное решение) График зависимости давления в пласте от расстояния для заданного момента времени (для нескольких моментов времени одновременно)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sc

```

```python
def p_wf_atm (r_m,t_day,k_mD=100, p_i_atm=200, q_s_m3day=50, b_m3m3=1.2, mu_cP=1.4,  h_m=10, m=0.4, c_t_1atm=1e-5):
  return p_i_atm - 18.41*q_s_m3day*b_m3m3*mu_cP/(k_mD*h_m)*(-1/2*sc.expi(-m*mu_cP*c_t_1atm*r_m**2/(0.00144*k_mD*t_day)))
```

```python
x = np.arange(0.1, 700, 10)
t = np.arange(0, 100, 1)
#n=np.arange(1,100,25)
#fig, (ax1, ax2) = plt.subplots(1,2)
#for i in n:
  #t_day=4*i
  #plt.plot(x, p_wf_atm(r_m,t_day)for r_m in x )
#fig,axes=plt.subplots(1,2)
#axes[0,1].plot(x, p_wf_atm(r_m=x,t_day=1), label='1 день')
plt.plot(x, p_wf_atm(r_m=x,t_day=1), label='1 день')
plt.plot(x, p_wf_atm(r_m=x,t_day=5), label='5 дней') 
plt.plot(x, p_wf_atm(r_m=x,t_day=50), label='50 дней')
plt.title('Воронка депрессии')
plt.xlabel('Расстояние, м')
plt.grid(which='major',
        color = 'r',
        linestyle = ':')
plt.ylabel('Давление на забое, атм')
plt.legend()
plt.show()
```

4. График изменения депрессии на пласт и логарифмической производной депрессии от времени в двойных логарифмических координатах.


```python
def depress_atm (t_day, r_m=0.1, k_mD=100, q_s_m3day=50, b_m3m3=1.2, mu_cP=1.4,  h_m=10, m=0.4, c_t_1atm=1e-5):
  return 18.41*q_s_m3day*b_m3m3*mu_cP/(k_mD*h_m)*(-1/2*sc.expi(-m*mu_cP*c_t_1atm*r_m**2/(0.00144*k_mD*t_day)))
```
